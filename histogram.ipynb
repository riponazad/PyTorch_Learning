```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
mu = 168 #mean
sigma = 5 #stddev
sample = 250
np.random.seed(0)
height_f = np.random.normal(mu, sigma, sample).astype(int)

mu = 176 #mean
sigma = 6 #stddev
sample = 250
np.random.seed(1)
height_m = np.random.normal(mu, sigma, sample).astype(int)
```

```python
print(height_f,height_m)
```

```python
gym = pd.DataFrame({'height_f': height_f, 'height_m': height_m})
gym
```

```python
gym.plot()
```

```python
gym.groupby('height_m').count().plot()
```

```python
gym.groupby('height_m').count().plot.bar()
```

```python
gym.hist()
```
